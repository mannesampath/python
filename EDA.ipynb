```python
#line plot:we use lineplot to find trend,ex:stockmarket
```

```python
dataset = sns.load_dataset("flights")
dataset.head()
```

```python
sns.lineplot(x='year',y='passengers',data=dataset)
```

```python
# moving average
```

```python
dataset['rolling_data']=dataset['passengers'].rolling(window=12).mean()
sns.lineplot(x='year',y='rolling_data',data=dataset,label = 'passenger')
sns.lineplot(x='year',y='passengers',data=dataset,label = 'rolling_avg')
plt.legend()
```
