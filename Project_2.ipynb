# FILLING THE NULL VALUES

```python
df = data.dropna()
```

```python
df.isna()
```

# CREATING NEW COLUMNS

```python
df['release_recent']=df['release_year'].apply(lambda x:'latest' if x >=2020 else 'recent' if x>=2010 else 'old')
print(df[['release_year','release_recent']].head(10))
```

# GROUP_BY AND AGG

```python
group_by = df.groupby('release_year').agg({'duration':['max','min']})
group_by.tail(10)
```

# PIVOTING TABALE

```python
df_pivot=df.pivot_table(index='release_year',columns='release_recent',values='duration',aggfunc='max')
df_pivot.tail(10)
```

# WHERE CONDITION

```python
data["show_idrange"]=np.where(data['show_id']> 's5','High','Low')
data.head(10)
```

# VISUALIZATION TECHNIQUES

```python
plt.figure(figsize=(10,5))
sns.lineplot(x='release_year',y= 'show_idrange',data=data)
plt.title('line_plot')
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.barplot(y='release_year',x= 'show_idrange',data=data)
plt.title("bar_plot")
plt.show()
```

```python

sns.scatterplot(x='release_year',y='duration',hue='type',data=data)
plt.title("scatter_plot")
plt.show()
```

# UNIVARIATE

```python
sns.histplot(data=data["release_year"],bins=15,kde=True)
plt.title("this is Histogram")
plt.show()
```
