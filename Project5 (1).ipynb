# CREATING NEW COLUMNS

```python
data['Price_Range']=data['Price'].apply(lambda x:'high' if x>30000 else 'medium' if x>20000 else 'low')
print(data[['Price','Price_Range']].head())
```

# GROUPBY AND AGG

```python
data_groupby = data.groupby('Brand').agg({'Price':['mean','min','max']})
data_groupby
```

# PIVOTING TABLE

```python
data_pivot = data.pivot_table(index='Brand',columns='Gearbox',values='Price',aggfunc='count')
data_pivot
```

# WHERE CONDITION

```python
data['product_condition']=np.where(data['Kilometers']>75000,'Old','Ne')
data.head()
```

# 2) How does the condition of a car (Status) affect its price?

# 3) Is there a correlation between mileage (Kilometers) and the price of a car?

```python
data1 = data['Kilometers'].corr(data['Price'])
data1=pd.DataFrame([[data1]],columns=['Corr between Kilometers and Price'])
sns.heatmap(data1,annot=True)
```
