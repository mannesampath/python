# CREATING NEW COLUMNS

```python
data['performance']=data['Likes'].apply(lambda x:'high' if x>30 else 'avg' if x>20 else 'low')
print(data[['Likes','performance']].head(10))
```

# GROUP_BY AND AGG

```python
data_groupby=data.groupby('Platform').agg({'Retweets':['mean','min','max']})
data_groupby
```

# PIVOTING TABLE

```python
pd.pivot_table=data.pivot_table(index='Platform',values='Likes',aggfunc='sum')
pd.pivot_table
```

# WHERE CONDITION

```python
data['Retweet_performance']=np.where(data['Retweets']>15,'High','Low')
data.head(10)
```

# VISUALIZATION TECHNIQUES

```python
sns.lineplot(x='Platform',y='Likes',data=data)
plt.title("this is a line plot")
plt.show()
```

```python
plt.figure(figsize=(15,5))
sns.barplot(x='Likes',y='Retweets',data = data)
plt.title("this is a bar plot")
plt.show()
```

```python
sns.scatterplot(x='Likes',y='Retweets',hue = 'Platform',data = data)
plt.title("this is a scatter plot")
plt.show()
```
