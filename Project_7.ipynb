# CREATING NEW COLUMNS

```python
data['Weights_range']=data['Weight (g)'].apply(lambda x:'High'if x>200 else 'Medium' if x>150 else 'Low')
print(data[['Weight (g)','Weights_range']].head())
```

# GROUPBY AND AGG

```python
data_groupby=data.groupby('Size (cm)').agg({'Weight (g)':['mean','min','max']})
data_groupby
```

# PIVOTING TABLE

```python
data_pivot=data.pivot_table(index='Size (cm)',values='Weight (g)',aggfunc='mean')
data_pivot
```

# WHERE CONDITION

```python
data['Product_Quality']=np.where(data['Quality (1-5)']>4,'High','Low')
data.head()
```

# QUESTIONS

# 1) Is there a correlation between the size (Size (cm)) and the weight (Weight (g)) of the fruit?

```python
data1 = data['Size (cm)'].corr(data['Weight (g)'])
data1=pd.DataFrame([[data1]],columns=['Corr between Size and Weight'])
sns.heatmap(data1,annot = True)
```

# 2) How does the sweetness (Brix) of the fruit correlate with its acidity (pH)? Do sweeter fruits have lower acidity?

```python
data2 = data['Brix (Sweetness)'].corr(data['pH (Acidity)'])
data2=pd.DataFrame([[data2]],columns = ['Corr between Brix and pH'])
sns.heatmap(data2,annot = True)
```

# 3) How does the softness (Softness (1-5)) of the fruit relate to its ripeness (Ripeness (1-5))?

```python
sns.scatterplot(x='Softness (1-5)',y='Ripeness (1-5)',data=data)
plt.title('scatterplot')
plt.show()
```

# 4) Do fruits with blemishes (Blemishes (Y/N)) tend to have lower ripeness or softness levels?

```python
plt.figure(figsize=(15,5))
sns.barplot(x='Blemishes (Y/N)',y='Softness (1-5)',data=data)
plt.title('barplot')
plt.show()
```

# 5) How do different fruit varieties (Variety) differ in terms of size, weight, sweetness, acidity, and ripeness?
